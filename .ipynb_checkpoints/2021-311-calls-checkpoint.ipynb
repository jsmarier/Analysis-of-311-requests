# Introduction

In this Jupyter notebook, I will be exploring the monthly service requests received by the City of Ottawa's 311 centre during the first three months of 2021. The version of the dataset used here was last updated on March 2, 2021. The latest version can be [found here]("https://open.ottawa.ca/datasets/2021-monthly-service-requests").

```python
# Import Pandas, Altair and the dataset.

import pandas as pd
import altair as alt

logs_311 = pd.read_excel("data/2021-311-calls.xlsx")
logs_311.head(20)
```

```python
logs_grouped = logs_311.groupby(by="SUBJECT").count()

logs_grouped = logs_grouped.reset_index()

logs_grouped
```

```python
plot_311 = alt.Chart(logs_grouped, width=500, height=300).mark_bar().encode(
    x=alt.X('SUBJECT:N', title="Department", sort="-y"),
    y=alt.Y('REASON:Q', title="Number of Requests")
).properties(title="Number of 311 requests received by the City of Ottawa between January and March 2021")

plot_311
```
